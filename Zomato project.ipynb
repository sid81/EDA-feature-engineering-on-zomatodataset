```python
 join_pd=pd.merge(df,df_country,on='Country Code')
join_pd
```

```python
join_pd.Country.value_counts()
```

```python
country_names=join_pd.Country.value_counts().index
country_names
```

```python
country_times=join_pd.Country.value_counts().values
country_times
```

```python
plt.pie(country_times[:3],labels=country_names[:3],autopct="'%1.2f%%'")
```

```python
##numerical values
```

```python
join_pd.columns
```

```python
join_pd.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
rating_finale=join_pd.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
rating_finale
```

```python
plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x= 'Aggregate rating',y="Rating count",data=rating_finale)
```

```python
sns.barplot(x= 'Aggregate rating',y="Rating count",hue='Rating color',data=rating_finale,palette=['blue','red','orange','yellow','green','green'])
```

## observation:-

1.Not rated count is vey high.
2.maximum number of rating are between 2.5 to 3.9(normal distribution curve)


## count plot-



```python
sns.countplot(x='Rating color',data=rating_finale,palette=['blue','red','orange','yellow','green','green'])
```
