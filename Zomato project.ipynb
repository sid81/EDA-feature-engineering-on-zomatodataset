```python
plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x= 'Aggregate rating',y="Rating count",data=rating_finale)
```

```python
sns.barplot(x= 'Aggregate rating',y="Rating count",hue='Rating color',data=rating_finale,palette=['blue','red','orange','yellow','green','green'])
```
